```python
import pandas as pd
bet = pd.read_csv('L_BET_1.csv', header=None)
skj = pd.read_csv('S_SKJ_1.csv', header=None)
```

```python
custom_headers = [
    "Gear", "Flag", "Year", "Region", "Annual-Catch-Estimate", 
    "Aggregate-Data-Area", "Aggregate-Data-Time", "Aggregate-Data-catch-recs", 
    "Operational-Data-catch-recs", "Length-Data-Area", "Length-Data-Time", 
    "Length-Data-Size-int", "Length-Data-Samples1", "Length-Data-Area", 
    "Length-Data-Time", "Length-Data-Size-int", "Length-Data-Samples2"
]

bet.columns = custom_headers
skj.columns = custom_headers
```

```python
import numpy as np

# List of columns to apply the transformation to
columns_to_convert = [
    'Annual-Catch-Estimate', 'Aggregate-Data-catch-recs', 
    'Operational-Data-catch-recs', 'Length-Data-Samples1', 'Length-Data-Samples2'
]

# Replace commas, then replace any non-numeric values with NaN and convert to integers
bet[columns_to_convert] = bet[columns_to_convert].replace({',': ''}, regex=True)

# Convert to numeric, setting errors='coerce' will turn non-numeric values into NaN
bet[columns_to_convert] = bet[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Optionally, you can fill NaNs with a default value, like 0
bet[columns_to_convert] = bet[columns_to_convert].fillna(0).astype(int)
bet.head()
```

```python
import numpy as np

# List of columns to apply the transformation to
columns_to_convert = [
    'Annual-Catch-Estimate', 'Aggregate-Data-catch-recs', 
    'Operational-Data-catch-recs', 'Length-Data-Samples1', 'Length-Data-Samples2'
]

# Replace commas, then replace any non-numeric values with NaN and convert to integers
skj[columns_to_convert] = skj[columns_to_convert].replace({',': ''}, regex=True)

# Convert to numeric, setting errors='coerce' will turn non-numeric values into NaN
skj[columns_to_convert] = skj[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Optionally, you can fill NaNs with a default value, like 0
skj[columns_to_convert] = skj[columns_to_convert].fillna(0).astype(int)
skj.head()
```
